```python
# prompt: data set import
import numpy as np
import pandas as pd

data = pd.read_csv('/content/parameters_data (1) (1).csv',index_col=0)

# Display the first few rows of the dataset
data
```

```python
# Convert to DataFrame
from scipy.stats import hmean, gmean
df = pd.DataFrame(data)

# Filter only numeric columns for calculations
numeric_columns = ["Winter", "Summer", "Rainy"]
def mean_deviation(values):
    mean_val = np.mean(values)
    return np.mean(np.abs(values - mean_val))

# Calculate statistics
df['Mean'] = df[numeric_columns].mean(axis=1)
df['Median'] = df[numeric_columns].median(axis=1)
df['Harmonic Mean'] = df[numeric_columns].apply(lambda row: hmean(row) if all(row > 0) else np.nan, axis=1)  # Harmonic Mean
df['Geometric Mean'] = df[numeric_columns].apply(lambda row: gmean(row) if all(row > 0) else np.nan, axis=1)  # Geometric Mean
df['Min'] = df[numeric_columns].min(axis=1)  # Minimum
df['Max'] = df[numeric_columns].max(axis=1)  # Maximum
df['Range'] = df['Max'] - df['Min']  # Range
df['Variance'] = df[numeric_columns].var(axis=1)  # Variance
df['STD'] = df['Variance']**0.5  # Standard Deviation
df['SE'] = df['STD'] / np.sqrt(3)  # Standard Error
df['MD'] = df[numeric_columns].apply(mean_deviation, axis=1)  # Mean Deviation
df['CV'] = (df['STD'] / df['Mean']) * 100  # Coefficient of Variation

# Display the updated DataFrame with new statistics
print(df[['Mean','Median', 'Harmonic Mean', 'Geometric Mean', 'Min', 'Max', 'Range', 'Variance', 'STD', 'SE', 'MD', 'CV']])
```

```python
# Extract only numerical columns
numeric_data = df[["Winter", "Summer", "Rainy"]]

transposed_data = numeric_data.T

correlation_matrix = transposed_data.corr()

print("Correlation Matrix:")
print(correlation_matrix)

transposed_data
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".4f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# prompt: find pca with eigen values

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming 'numeric_data' DataFrame from the previous code is available
# If not, create it from your data

# Standardize the data
x = StandardScaler().fit_transform(numeric_data)

# Calculate the covariance matrix
covariance_matrix = np.cov(x, rowvar=False)

# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and eigenvectors in descending order
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]


# Explained variance ratio
explained_variance_ratio = eigenvalues / eigenvalues.sum()

# Choose the number of principal components (e.g., top 2)
n_components = 4

# Project the data onto the principal components
principal_components = x.dot(eigenvectors[:, :n_components])


print("Eigenvalues:", eigenvalues)
print("Eigenvectors:", eigenvectors)
print("Explained Variance Ratio:", explained_variance_ratio)
print("Principal Components:", principal_components)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define the pairs and their labels
pairs = [
    ("pH", "Cl(mg/L)"),
    ("Cl(mg/L)", "BOD(mg/L)"),
    ("BOD(mg/L)", "TDS(mg/L)"),
    ("TDS(mg/L)", "WT(°C)"),
    ("WT(°C)", "TH(mg/L)")
]

# Loop over each pair and perform linear regression
for param2, param1 in pairs:
    # Prepare the data for linear regression
    X = transposed_data[param1].values.reshape(-1, 1)  # Independent variable (param1)
    y = transposed_data[param2].values  # Dependent variable (param2)

    # Create and train the linear regression model
    model = LinearRegression()
    model.fit(X, y)

    # Make predictions
    y_pred = model.predict(X)

    # Plot the regression line
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, color='blue', label='Actual values')  # Actual values
    plt.plot(X, y_pred, color='red', label='Regression Line')  # Predicted values (Regression line)
    plt.xlabel(param1)
    plt.ylabel(param2)
    plt.title(f"Linear Regression: {param2} vs {param1}")
    plt.legend()
    plt.show()

    # Print the model's coefficients and intercept
    print(f"Regression for {param2} ~ {param1}:")
    print(f"Intercept: {model.intercept_:.4f}")
    print(f"Coefficient: {model.coef_[0]:.4f}")

    # Evaluate the model
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"R-squared (R²): {r2:.4f}\n")
```

```python
from sklearn.model_selection import cross_val_score

# Example for linear regression
model = LinearRegression()
cv_scores = cross_val_score(model, X, y, cv=3, scoring='neg_mean_squared_error')
print(f"Cross-validated MSE: {-cv_scores.mean()}")
```
